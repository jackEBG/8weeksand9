# 	Merge Datasets for Geolocation Analysis

```python
import socket
import struct
# Function to convert IP to integer
def ip_to_int(ip):
    return struct.unpack("!I", socket.inet_aton(ip))[0]

# Load datasets
fraud_data = pd.read_csv(r'C:\Users\Admin.DESKTOP-M4R2VLU\8weeksand9\data\Fraud_clean_data.csv')
ip_data = pd.read_csv(r'C:\Users\Admin.DESKTOP-M4R2VLU\8weeksand9\data\IpAddress_clean2_Country.csv')
creditcard_data = pd.read_csv(r'C:\Users\Admin.DESKTOP-M4R2VLU\8weeksand9\data\creditcard.csv')
```

```python
# Convert IP addresses to integers
fraud_data['ip_as_int'] = fraud_data['ip_address'].apply(ip_to_int)
ip_data['lower_bound_ip'] = ip_data['lower_bound_ip_address'].apply(ip_to_int)
ip_data['upper_bound_ip'] = ip_data['upper_bound_ip_address'].apply(ip_to_int)
```

```python
# Merge datasets
merged_data = fraud_data.merge(ip_data, how='left', left_on='ip_as_int', right_on='lower_bound_ip')
merged_data = merged_data[(merged_data['ip_as_int'] >= merged_data['lower_bound_ip']) & 
                           (merged_data['ip_as_int'] <= merged_data['upper_bound_ip'])]
```

```python
merged_data
```

```python
# Feature Engineering
transaction_frequency = fraud_data.groupby('user_id')['purchase_value'].count().reset_index()
transaction_frequency.columns = ['user_id', 'transaction_frequency']
fraud_data['purchase_time'] = pd.to_datetime(fraud_data['purchase_time'])
fraud_data['time_diff'] = fraud_data['purchase_time'].diff().dt.total_seconds()
transaction_velocity = fraud_data.groupby('user_id')['time_diff'].mean().reset_index()
transaction_velocity.columns = ['user_id', 'average_transaction_velocity']
features = transaction_frequency.merge(transaction_velocity, on='user_id')
```

```python
features
```

```python
# Time-Based Features
fraud_data['hour_of_day'] = fraud_data['purchase_time'].dt.hour
fraud_data['day_of_week'] = fraud_data['purchase_time'].dt.dayofweek
```

```python
# Normalization and Scaling
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features[['transaction_frequency', 'average_transaction_velocity']])
features[['scaled_frequency', 'scaled_velocity']] = features_scaled
```

```python

# Encode Categorical Features
encoded_features = pd.get_dummies(merged_data[['source', 'browser', 'sex', 'country']], drop_first=True)
final_data = pd.concat([merged_data, encoded_features], axis=1)
final_data = final_data.drop(columns=['Unnamed: 0']).reset_index(drop=True)
```
